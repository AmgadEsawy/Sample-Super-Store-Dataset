```python
# new DataFrame : 
df1 = df[['Ship Mode','Segment', 'City', 'Category','Sub-Category','State','Region', 'Quantity','Discount','Sales','Profit']]
df1.head()
```

```python
# total quantity, sales, and profit per category
df1.groupby('Category')['Quantity','Sales','Profit'].sum()
```

```python
# mean, min, and max sales by Category
df1.groupby('Category')['Sales'].agg([np.mean, np.min, np.max])
```

```python
# mean, min, and max profit by category
df1.groupby('Category')['Profit'].agg([np.mean, np.min, np.max])
```

```python
# mean, min, and max profit by region
df1.groupby('Region')['Profit'].agg([np.mean, np.min, np.max])
```

```python
#pairwise relationships in df1
g = sns.pairplot(df1, hue = 'Sub-Category')
```

```python
# distribution plot
g = sns.distplot(df1['Sales'], bins = 100, rug = True, kde_kws = {'shade' : True})
g.set_title('Distribution Plot')
```

```python
# top 10 cities 
top_10 = df1.sort_values(by = 'Sales', ascending = False, ignore_index = True)
top_10.head(10)
```

```python
top_10_city_sales = top_10.iloc[:10, :]
top_10_city_sales
```

### Top 10 city sales per category :

```python
g = sns.barplot(x = 'City', y = 'Sales', data = top_10_city_sales, hue = 'Category')
plt.xticks(rotation = 90)
g.set_title('Top 10 city sales per category')
```

### Top 10 city sales per sub-category :

```python
g = sns.barplot(x = 'City', y = 'Sales', data = top_10_city_sales, hue = 'Sub-Category')
plt.xticks(rotation = 90)
g.set_title('Top 10 city sales per Sub-Category')
```
